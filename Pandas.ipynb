# Analíse Exploratoria Demanda


```python
dt.head()
```

```python
dt['Valor'].mean()
```

```python
dt.groupby('Tipo')['Valor'].mean(numeric_only= True)
```

```python
grafico = dt.groupby('Tipo')[['Valor']].mean().reset_index().sort_values('Valor')
grafico.plot(kind='bar', title='Valor Médio por Tipo',figsize=(10,6),color='purple', x='Tipo', y='Valor')
```

```python
dt.Tipo.unique()
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala',
                      'Prédio Inteiro', 'Loja/Salão',
                      'Galpão/Depósito/Armazém',
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']
```

```python
dt.query('@imoveis_comerciais in Tipo')
```

```python
dtc = dt.query('@imoveis_comerciais not in Tipo')
```

```python
dtc.head()
```

```python
dtc.Tipo.unique()
```

```python
grafico = dtc.groupby('Tipo')[['Valor']].mean().reset_index().sort_values('Valor')
grafico.plot(kind='bar', title='Valor Médio por Tipo',figsize=(10,6),color='purple', x='Tipo', y='Valor')
```

```python
dtc.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')
```

```python
df_percentual_tipo =dtc.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')
```

```python
df_percentual_tipo.plot(kind='bar', title='Percentual por Tipo', color= 'red', figsize=(10,6), xlabel='Tipo', ylabel='Valor')
```
